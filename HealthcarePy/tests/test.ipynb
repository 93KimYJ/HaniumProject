```python
import pandas as pd

life = pd.read_csv("https://raw.githubusercontent.com/dongupak/DataML/main/csv/life_expectancy.csv")
life
```

```python
import seaborn as sns
import matplotlib as plt

sns.set(rc={'figure.figsize':(22, 20)})
correlation_matrix = life.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
import numpy as np
print("기대수명과의 상관계수")
print(life.corr().round(3)['Life expectancy'])
```

```python
c = life.corr().round(2)['Life expectancy']
c = np.abs(c)
print(c.sort_values(ascending=False)[1:8])
```

```python
sns.pairplot(life[['Life expectancy', 'Schooling', 'Income composition of resources', 'Adult mortality', 'HIV/AIDS', 'Thinness 1-19 years']])
```

```python
life[['Schooling', 'Income composition of resources', 'Adult mortality', 'HIV/AIDS', 'Thinness 1-19 years']].isna().sum()
```

```python
life.dropna(inplace=True)
X = life[['Schooling', 'Income composition of resources', 'Adult mortality', 'HIV/AIDS', 'Thinness 1-19 years']]
y = life['Life expectancy']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

regr = LinearRegression()
regr.fit(X_train, y_train)
print("선형 회귀 모델의 점수 =", regr.score(X_test, y_test).round(3))
```

```python
life.dropna(inplace=True)
X = life[['Year', 'Adult mortality', 'Infant deaths', 'Alcohol', 'Percentage expenditure', 'Hepatitis B', 'Measles', 'BMI', 'Under-five deaths'
          , 'Polio', 'Total expenditure', 'Diphtheria', 'HIV/AIDS', 'GDP', 'Population', 'Thinness 1-19 years', 'Thinness 5-9 years', 'Income composition of resources', 'Schooling']]
y = life['Life expectancy']
```
